#### Using BeautifulSoup

```python
d = {"Company":[], "Name":[], "Inches":[], "Screen_Resolution":[],"CPU":[], "RAM":[],"Memory":[],"GPU":[],"OpSys":[],"Weights":[], "Price":[]}

for i in range(1,2):
    # if i == 1: 
    url = f'https://www.newegg.com/p/pl?N=100006740%204131%20601183898%204814%204841&page={i}'
    # HTTP Request
    webpage = requests.get(url)

    # Soup Object containing all data
    soup = BeautifulSoup(webpage.content, "html.parser")

    links_to_details = soup.find_all('a', class_=['item-title', 'L-Affiliate-Tagged'])
    print(f'page {i} has {len(links_to_details)} items')
    # Loop for extracting product  details from each link 
    for link in links_to_details:
        new_webpage = requests.get(link.get('href'))

        new_soup = BeautifulSoup(new_webpage.content, "html.parser")
        
        th_tags = new_soup.find_all('th')
        company = ''
        price = ''
        nches ='',
        screenResolution=''
        cpu=''
        gpu=''
        ram=''
        memory='' 
        opsys=''
        weight=''
        price=''
        name = new_soup.find("h1", class_ = 'product-title').text
        price = new_soup.find('li', class_= 'price-current').text
        for th in th_tags: 
            key = th.text
            if key.rstrip() == 'Brand':
                company = th.find_next_sibling("td").text
            if key.rstrip() == 'Screen Size':
                inches = th.find_next_sibling("td").text
            if key.rstrip() == 'Resolution':
                screenResolution = th.find_next_sibling("td").text
            if key.rstrip() == 'CPU Type':
                cpu = th.find_next_sibling("td").text
            if key.rstrip() == 'Memory':
                ram = th.find_next_sibling("td").text
            if key.rstrip() == 'SSD' or key.rstrip() == 'HDD':
                memory = th.find_next_sibling("td").text
            if key.rstrip() == 'GPU/VPU':
                gpu = th.find_next_sibling("td").text
            if key.rstrip() == 'Dimensions & Weight' or key.rstrip() == "Weight":
                weight = th.find_next_sibling("td").text
            if key.rstrip() == 'Operating System':
                opsys = th.find_next_sibling("td").text
                    
        d['Company'].append(company)
            
        d['Name'].append(name)
        
        d['Inches'].append(inches)
        
        d['Screen_Resolution'].append(screenResolution)
        
        d['CPU'].append(cpu)
        
        d['RAM'].append(ram)
        
        d['Memory'].append(memory)
        
        d['GPU'].append(gpu)
        
        d['OpSys'].append(opsys)
        
        d['Weights'].append(weight)
        
        d['Price'].append(price)

```

```python
newegg_df = pd.DataFrame.from_dict(d)
newegg_df.head(10)
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
d = {"Company":[], "Name":[], "Inches":[], "Screen_Resolution":[],"CPU":[], "RAM":[],"Memory":[],"GPU":[],"OpSys":[],"Weights":[], "Price":[]}

# Kh·ªüi t·∫°o tr√¨nh duy·ªát
driver = webdriver.Chrome("chromedriver.exe")

# M·ªü trang web
for i in range(1,2):
    url = f"https://www.newegg.com/p/pl?N=100167750%204131%204841&page={i}"
    driver.get(url)
    wait = WebDriverWait(driver, 10)
    wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.item-cells-wrap")))
    links = driver.find_elements(By.CSS_SELECTOR, 'a.item-title')
    link_to_details = [link.get_attribute('href') for link in links]
    print(f'page {i} has {len(link_to_details)} items')
    if len(link_to_details) ==0 :
        continue
    else: 
        for link in link_to_details: 
            driver.get(link)
            th_tags = driver.find_elements(By.CSS_SELECTOR, 'th')
            company = ''
            price = ''
            inches ='',
            screenResolution=''
            cpu=''
            gpu=''
            ram=''
            memory='' 
            opsys=''
            weight=''
            price=''
            name = driver.find_element(By.CSS_SELECTOR, 'h1.product-title').text
            price = driver.find_element(By.CSS_SELECTOR, 'li.price-current').text
            for th in th_tags: 
                key = th.text
                if key.rstrip() == 'Brand':
                    company = th.find_element(By.XPATH, "following-sibling::td").text
                if key.rstrip() == 'Screen Size':
                    inches = th.find_element(By.XPATH, "following-sibling::td").text
                if key.rstrip() == 'Resolution':
                    screenResolution = th.find_element(By.XPATH, "following-sibling::td").text
                if key.rstrip() == 'CPU Type':
                    cpu = th.find_element(By.XPATH, "following-sibling::td").text
                if key.rstrip() == 'Memory':
                    ram = th.find_element(By.XPATH, "following-sibling::td").text
                if key.rstrip() == 'SSD' or key.rstrip() == 'HDD':
                    memory = th.find_element(By.XPATH, "following-sibling::td").text
                if key.rstrip() == 'GPU/VPU':
                    gpu = th.find_element(By.XPATH, "following-sibling::td").text
                if key.rstrip() == 'Dimensions & Weight' or key.rstrip() == "Weight":
                    weight = th.find_element(By.XPATH, "following-sibling::td").text
                if key.rstrip() == 'Operating System':
                    opsys = th.find_element(By.XPATH, "following-sibling::td").text
                        
            d['Company'].append(company)
                
            d['Name'].append(name)
            
            d['Inches'].append(inches)
            
            d['Screen_Resolution'].append(screenResolution)
            
            d['CPU'].append(cpu)
            
            d['RAM'].append(ram)
            
            d['Memory'].append(memory)
            
            d['GPU'].append(gpu)
            
            d['OpSys'].append(opsys)
            
            d['Weights'].append(weight)
            
            d['Price'].append(price)
driver.quit()
```

```python
import pandas as pd
import matplotlib as plt
from bs4 import BeautifulSoup
import requests
import time
my_list = [
    {"https://www.newegg.com/p/pl?Submit=StoreIM&Category=223&page=": 5},
    {"https://www.newegg.com/p/pl?N=100006740%204131%204814%20600364045&page=": 24},
    {"https://www.newegg.com/p/pl?N=100006740%20600004970%20600364421%20600566284%20600472974%20601395534%204131%204841&page=": 26},
    {'https://www.newegg.com/p/pl?N=100006740%204131%20601183898%204814%204841&page=':3},
    {'https://www.newegg.com/p/pl?N=100167750%204131%204841&page=':4},
]

d = {"Company":[], "Name":[], "Inches":[], "Screen_Resolution":[],"CPU":[], "RAM":[],"Memory":[],"GPU":[],"OpSys":[],"Weights":[], "Price":[]}

# Iterate over the list
for item in my_list:
    # Iterate over the key-value pairs of each dictionary
    for key, value in item.items():
        # find all pages of url
        for i in range(1, value):
            url = f"{key}{i}"
            webpage = requests.get(url)

            # Soup Object containing all data
            soup = BeautifulSoup(webpage.content, "html.parser")

            links_to_details = soup.find_all('a', class_=['item-title', 'L-Affiliate-Tagged'])
            print(f'page {i} of url {key} has {len(links_to_details)} items')
            # Loop for extracting product  details from each link 
            for link in links_to_details:
                new_webpage = requests.get(link.get('href'))

                new_soup = BeautifulSoup(new_webpage.content, "html.parser")
                
                th_tags = new_soup.find_all('th')
                company = ''
                price = ''
                nches ='',
                screenResolution=''
                cpu=''
                gpu=''
                ram=''
                memory='' 
                opsys=''
                weight=''
                price=''
                name = new_soup.find("h1", class_ = 'product-title').text
                price = new_soup.find('li', class_= 'price-current').text
                for th in th_tags: 
                    key = th.text
                    if key.rstrip() == 'Brand':
                        company = th.find_next_sibling("td").text
                    if key.rstrip() == 'Screen Size':
                        inches = th.find_next_sibling("td").text
                    if key.rstrip() == 'Resolution':
                        screenResolution = th.find_next_sibling("td").text
                    if key.rstrip() == 'CPU Type':
                        cpu = th.find_next_sibling("td").text
                    if key.rstrip() == 'Memory':
                        ram = th.find_next_sibling("td").text
                    if key.rstrip() == 'SSD' or key.rstrip() == 'HDD':
                        memory = th.find_next_sibling("td").text
                    if key.rstrip() == 'GPU/VPU':
                        gpu = th.find_next_sibling("td").text
                    if key.rstrip() == 'Dimensions & Weight' or key.rstrip() == "Weight":
                        weight = th.find_next_sibling("td").text
                    if key.rstrip() == 'Operating System':
                        opsys = th.find_next_sibling("td").text
                            
                d['Company'].append(company)
                    
                d['Name'].append(name)
                
                d['Inches'].append(inches)
                
                d['Screen_Resolution'].append(screenResolution)
                
                d['CPU'].append(cpu)
                
                d['RAM'].append(ram)
                
                d['Memory'].append(memory)
                
                d['GPU'].append(gpu)
                
                d['OpSys'].append(opsys)
                
                d['Weights'].append(weight)
                
                d['Price'].append(price)
            time.sleep(3)



```
